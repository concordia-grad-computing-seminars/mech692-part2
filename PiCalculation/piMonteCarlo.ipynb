# My first Monte-Carlo Simulation

This notebook calculates an approximate value for the constant pi, based on the selection of a random number of points.

Select the number of points:

```octave
npoints = 1000
```

```octave
inPoints = 0;

inPointsArr  = [];
outPointsArr = [];

for i = 1:npoints
  xPos = 2*rand()-1;
  yPos = 2*rand()-1;
  rPos = sqrt(xPos^2 + yPos^2);
  if (rPos <= 1)
    inPoints = inPoints + 1;
    inPointsArr  = [inPointsArr; xPos, yPos];
  else
    outPointsArr = [outPointsArr; xPos, yPos];  
  end  
end
value = 4*inPoints/npoints
```

```octave
plot(inPointsArr(:,1),inPointsArr(:,2),'b.',outPointsArr(:,1),outPointsArr(:,2),'r.')
```
